```python
import xarray as xr
tech = "onwind"
ds = xr.open_dataset(f"resources/{run_name}/profile_1_{tech}.nc")
# Average CF per bus and bin
average_cf = ds["profile"].mean(dim="time")

# Select CF for this bus. Use a new variable to avoid overwriting.
bus_cf = average_cf.sel(bus='DE0 0')

print("Bus CF before extracting values:")
print(bus_cf)

# Extract values. Assuming (year, bin)
if "year" in bus_cf.dims:
    cf_values = bus_cf.isel(year=0).values
else:
    cf_values = bus_cf.values

print("CF values for bus 'DE0 0':")
print(cf_values)

# Distribute shares proportional to average capacity factors
shares = cf_values / cf_values.sum()
sum_shares = shares.sum()
shares = shares / sum_shares

print("Sum of shares after normalization:")
print(shares.sum())
```
