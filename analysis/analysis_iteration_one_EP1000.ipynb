```python
import pandas as pd

def get_avg_cf(gens_idx):
    # Intersection with profiles
    valid = gens_idx.intersection(n.generators_t.p_max_pu.columns)
    if valid.empty:
        return pd.Series(dtype=float)
    return n.generators_t.p_max_pu.loc[:, valid].mean()

onwind_gens = n.generators[
        n.generators.carrier.isin(['onwind']) & n.generators.p_nom_extendable
    ]
TARGET_ONWIND_MW = 160000.0
cf_onwind = get_avg_cf(onwind_gens.index)
total_cf_onwind = cf_onwind.sum()
# shares_onwind = cf_onwind / total_cf_onwind
p_nom_max = n.generators.loc[onwind_gens.index, "p_nom_max"].fillna(0)
# Weight share by (CF * Technical Potential) to ensure cap scales with available land
weights_onwind = cf_onwind * p_nom_max
total_weight_onwind = weights_onwind.sum()
shares_onwind = weights_onwind / total_weight_onwind
alloc_onwind = shares_onwind * TARGET_ONWIND_MW
print(alloc_onwind)
common_idx = alloc_onwind.index
print(common_idx)
print(n.generators.loc[common_idx, 'p_nom_max'])

solar_hsat_gens = n.generators[
        n.generators.carrier.isin(['solar-hsat']) & n.generators.p_nom_extendable
    ]
TARGET_SOLAR_HSAT_MW = 400000.0
cf_solar_hsat = get_avg_cf(solar_hsat_gens.index)
total_cf_solar_hsat = cf_solar_hsat.sum()
p_nom_max = n.generators.loc[solar_hsat_gens.index, "p_nom_max"].fillna(0)
# Weight share by (CF * Technical Potential) to ensure cap scales with available land
weights_solar_hsat = cf_solar_hsat * p_nom_max
total_weight_solar_hsat = weights_solar_hsat.sum()
shares_solar_hsat = weights_solar_hsat / total_weight_solar_hsat
alloc_solar_hsat = shares_solar_hsat * TARGET_SOLAR_HSAT_MW
print(alloc_solar_hsat)
common_idx = alloc_solar_hsat.index
print(common_idx)
print(n.generators.loc[common_idx, 'p_nom_max'])
```
